```python
import os
import openpyxl
import pandas as pd

# Lambda 형 활용!! (중요)
match_list = list(filter(lambda x: "토지"in x, os.listdir('./')))



# file_list = os.listdir('./')

# # 리스트형
# print(file_list)

# match_list = list()

# for word in file_list:
#     if '토지' in word:
#         match_list.append(word)

match_list
```

```python
df = pd.DataFrame()
# print(type(dff))

max_row = 0

for i in match_list:

    wb = openpyxl.load_workbook('./' + i, data_only= True)

    # 현재 활성화된 시트 가져오기
    ws = wb.active

    df_temp = pd.DataFrame(ws.values)

    # 데이터프레임
    # print(type(df.loc[~df[1].isnull()]))

    df_temp = df_temp.loc[~df_temp[1].isnull()]

    # 제외된 인덱스 기준이다 (0 = 15)
    df_temp.columns = df_temp.iloc[0].tolist()

    df_temp.drop([16], axis=0, inplace=True)

    df_temp.reset_index(inplace=True, drop=True)
    
    print(i + ': ' + str(len(df_temp.index)))
    
    max_row = max_row + len(df_temp.index)

    df = pd.concat([df, df_temp], axis=0)

print(max_row)
    
df
```

```python
df.to_csv('df_union.csv', index=False, encoding='euc-kr')
```
