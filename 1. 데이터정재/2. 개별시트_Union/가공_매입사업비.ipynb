```python
import pandas as pd
import openpyxl

# 여러 줄 한번에 output 하기
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

wb = openpyxl.load_workbook('C:/Users/dfocus/PycharmProjects/pythonProject/jupyter/test_excel/test.xlsx', data_only= True)

ws = wb.get_sheet_by_name("매입사업비")

df = pd.DataFrame(ws.values)

df.drop([0,1,2], axis=0, inplace=True)
# df.drop([0], axis=1, inplace=True)

df.dropna(how='all', axis = 0, inplace=True)
df.dropna(how='all', axis = 1, inplace=True)

df.reset_index(inplace=True, drop=True)

df.head(8)
```

```python
a = []

a.append(df.iloc[0,0])
a.append(df.iloc[0,1])
a.append('사업비')
a.append(df.iloc[0,3])
a.append(df.iloc[0,4].replace("’", ''))


max_columns = len(df.columns)

for i in range(5, max_columns):
    
    # null 값 체크 (NA -> TRUE)
    if df.iloc[1].notna()[i] :
        a.append(str(df.iloc[1,i-1]) + '_' + df.iloc[2,i] )
        
    else: 
        a.append(df.iloc[1,i] + '_' + df.iloc[2,i] )
        

print(a)

df.columns= a

df.drop([0,1,2], axis=0, inplace=True)

df.drop(['20년 월별추진계획'], axis=1, inplace=True)

df.reset_index(inplace=True, drop=True)

df.head(8)
```

```python
df.to_csv('df_output.csv', index=False, encoding='euc-kr')
```
