```python
# https://tjansry354.tistory.com/9
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = 'C:/Users/dfocus/PycharmProjects/pythonProject/jupyter/1.train_test/titanic_train.csv'
train_df = pd.read_csv(file_path, sep=',')
train_df
```

```python
#  결측값 확인

# 1. pclass : 티켓 등급
# 2. survived : 생존 유무(생존=1, 죽음=0)
# 3. name : 승객 이름
# 4. sex : 성별
# 5. age : 나이
# 6. sibsp : 동승자 수(형제 또는 배우자)
# 7. parch : 동승자 수(부모 또는 자녀)
# 8. ticket : 티켓 번호
# 9. far : 티켓 요금
# 10. cabin : 선실 번호
# 11. embarked : 탑승 장소(선착장)

train_df.isnull().sum()
```

```python
# 의미없는 칼럼 제외 -> asix=0 행방향 & asix=1 열방향
train_df=train_df.drop(['cabin','body','home.dest'], axis=1)
```

```python
# replace로 name split 일관성 수정 -> first_name 가져오기
first_name = pd.DataFrame(train_df['name'].str.replace('.',','))['name'].str.split(',').str.get(1)

train_df['first_name'] = first_name.str.replace(' ', '')

# print(type(first_name))

train_df
```

```python
# null값은 평균구할때 포함되지 않는 확인 테스트
# https://blog.naver.com/PostView.nhn?blogId=wideeyed&logNo=221578773214&parentCategoryNo=&categoryNo=50&viewDate=&isShowPopularPosts=false&from=postView

train_age_avg_df = train_df[['first_name', 'sex', 'age']]

# 한번에 2번 group by 하기
test1 = train_age_avg_df.groupby(by=['first_name', 'sex']).agg(['mean', 'count'])

# multiple index 가공
# https://stackoverflow.com/questions/54828684/pandas-multi-index-issue-in-merging-dataframes

test2 = (train_age_avg_df.groupby(by=['first_name', 'sex'])['age']
                        .agg([('mean_return', 'mean'), ('cnt_return', 'count')])).reset_index()

# 내림차순 정렬하기
test2.sort_values(by=['cnt_return'], axis = 0, ascending=False) 
```

```python
# 기준 dataframe에서 group by 칼럼 추가
# https://rfriend.tistory.com/403

train_df['avg_count'] = train_df.groupby(by=['first_name', 'sex'], as_index=False).age.transform('mean')

train_df
```

```python
num=0

for i in train_df['age'] :
    if train_df['age'].isna()[num] :
        train_df['age'][num] = train_df['avg_count'][num].round(0)
    num+=1

# 반올림
train_df.age = train_df.age.round(2)

train_df=train_df.drop(['avg_count'], axis=1)

train_df
```

```python
train_df['embarked'].value_counts()
```

```python
num=0
for i in train_df['embarked'] :
    if train_df['embarked'].isna()[num]  :
        train_df['embarked'][num]='S'
    num+=1
    

train_df.isnull().sum()
```

```python
# 나이
train_df.loc[train_df['age']<=20, 'age']=0
train_df.loc[(train_df['age']<=40)&(train_df['age']>20), 'age']=1
train_df.loc[(train_df['age']<=60)&(train_df['age']>40), 'age']=2
train_df.loc[(train_df['age']<=80)&(train_df['age']>60), 'age']=3
train_df.loc[train_df['age']>80, 'age']=40

# 자리
train_df.loc[train_df['embarked']=='S', 'embarked']=1
train_df.loc[train_df['embarked']=='C', 'embarked']=2
train_df.loc[train_df['embarked']=='Q', 'embarked']=3

# 성별
train_df.loc[train_df['sex']=='female', 'sex']=0
train_df.loc[train_df['sex']=='male', 'sex']=1

# 이름 (많은 이름 순 3위까지 -> 나머지는 Others)
train_df.loc[train_df['first_name']=='Mr', 'name']=1
train_df.loc[train_df['first_name']=='Miss', 'name']=2
train_df.loc[train_df['first_name']=='Mrs', 'name']=3

train_df.loc[~train_df['first_name'].isin(['Miss', 'Mr', 'Mrs']), 'name'] = 4

train_df=train_df.drop(['first_name'], axis=1)

# 가족
train_df['family']=train_df['sibsp']+train_df['parch']
train_df=train_df.drop(['sibsp','parch'], axis=1)

train_df.info()
```

```python
# 형 변환
train_df = train_df.astype({'name': 'int'})
train_df = train_df.astype({'sex': 'int'})
train_df = train_df.astype({'embarked': 'int'})


corr_df = train_df.corr(method='pearson')

corr_df


# train_df.to_csv('train_df_test.csv', index=False, encoding='euc-kr')
```
