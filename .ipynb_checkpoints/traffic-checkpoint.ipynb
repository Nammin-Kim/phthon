```python
# https://tjansry354.tistory.com/9
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = 'C:/Users/dfocus/PycharmProjects/pythonProject/jupyter/titanic_train.csv'
train_df = pd.read_csv(file_path, sep=',')
train_df
```

```python
#  결측값 확인
train_df.isnull().sum()
```

```python
# 의미없는 칼럼 제외 -> asix=0 행방향 & asix=1 열방향
train_df=train_df.drop(['cabin','body','home.dest'], axis=1)

# replace로 name split 일관성 수정
test_df = pd.DataFrame(train_df['name'].str.replace('.',','))

test_df
```

```python
# warring 관련 내용 (http://blog.naver.com/PostView.nhn?blogId=wideeyed&logNo=221817400937)
pd.set_option('mode.chained_assignment',  None)


# name칼럼 스플릿
name1 = test_df['name'].str.split(',').str.get(0)
name2 = test_df['name'].str.split(',').str.get(1)
name3 = test_df['name'].str.split(',').str.get(2)

test_df['name1'] = name1
test_df['name2'] = name2
test_df['name3'] = name3

train_df['first_name'] = name2

display(test_df, train_df)
```

```python
# null값은 평균구할때 포함되지 않는 확인 테스트
# https://blog.naver.com/PostView.nhn?blogId=wideeyed&logNo=221578773214&parentCategoryNo=&categoryNo=50&viewDate=&isShowPopularPosts=false&from=postView

test_df = train_df[['first_name', 'sex', 'age']]
test_df_fix =test_df.dropna(axis=0)

test_df = test_df.groupby(by=['first_name', 'sex'], as_index=False).mean()

test_df_fix = test_df_fix.groupby(by=['first_name', 'sex'], as_index=False).mean()

# https://m.blog.naver.com/PostView.nhn?blogId=youji4ever&logNo=221643578115&proxyReferer=https:%2F%2Fwww.google.com%2F
test_df.columns = ['first_name', 'sex', 'age_mean']

display(test_df,test_df_fix)
```

```python
# join 후 추가

train_df = pd.merge(train_df, test_df, on = ['first_name', 'sex'], how='inner')

train_df
```
